# Data Cleaning

### Modify the data type

```python
# Convert non-numeric values to numeric values in the BasePay, OvertimePay, OtherPay,Benefits columns.
Salaries['BasePay'] = pd.to_numeric(Salaries['BasePay'], errors='coerce')
Salaries['OvertimePay'] = pd.to_numeric(Salaries['OvertimePay'], errors='coerce')
Salaries['OtherPay'] = pd.to_numeric(Salaries['OtherPay'], errors='coerce')
Salaries['Benefits'] = pd.to_numeric(Salaries['Benefits'], errors='coerce')
```

```python
# Normalizing job titles: strip whitespace and convert to lowercase
Salaries['JobTitle'] = Salaries['JobTitle'].str.strip().str.lower()
```

### Handle Missing Values

```python
Salaries['Benefits'].fillna(0.0, inplace=True)
Salaries['BasePay'].fillna(0.0, inplace=True)
```

```python
# Drop the 'Notes' column
Salaries.drop(columns=['Notes'], inplace=True, errors='ignore')
```

### Removing data where EmployeeName and his details is not provided

```python
Salaries = Salaries[Salaries['EmployeeName'] != "Not provided"]
```

### Status Column
#### Group the DataFrame by 'JobTitle' and find the mode (most common value) of the 'Status' column within each group.
#### If a mode is found for a particular 'JobTitle', fill the NaN values in the 'Status' column for that group with the mode value.
#### If no mode is found for a specific 'JobTitle', fill the NaN values in the 'Status' column with the overall mode (most common value) of the entire 'Status' column.

```python
# Find mode of 'Status' for each 'JobTitle'
mode_per_job_title = Salaries.groupby('JobTitle')['Status'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None)

# Find overall mode of 'Status'
overall_mode = Salaries['Status'].mode().iloc[0]

# Fill NaN values in 'Status' column with mode of 'Status' for each 'JobTitle'
Salaries['Status'] = Salaries.apply(lambda row: mode_per_job_title[row['JobTitle']] if pd.isnull(row['Status']) and row['JobTitle'] in mode_per_job_title else row['Status'], axis=1)

# Fill remaining NaN values in 'Status' column with overall mode of 'Status'
Salaries['Status'].fillna(overall_mode, inplace=True)
Salaries.head(10)
```
